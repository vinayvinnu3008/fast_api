````python
# Day 1 practice

# Example 1: variables
name = "Alice"
age = 25
is_student = True

print(name)
print(age)
print(is_student)

# Example 2: input
name = input("Enter your name: ")
print("Hello", name)

# Day 2 practice

# Example 3: list and loop
students = ["Asha", "Ravi", "Nina"]
for student in students:
    print(student)

# Example 4: dictionary
profile = {"name": "Asha", "age": 20}
print(profile["name"])
print(profile["age"])

# Example 5: condition
if profile["age"] >= 18:
    print("Adult")
else:
    print("Minor")
````
